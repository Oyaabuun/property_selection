```python
from pymongo import MongoClient
from bson import ObjectId
from datetime import datetime
```

```python
MONGO_URI = "mongodb://127.0.0.1:27017"
DB_NAME = "propertyai"

client = MongoClient(MONGO_URI)
db = client[DB_NAME]

print("Connected to DB:", db.name)
```

```python
properties = [
    {
        "title": "2BHK near KIIT",
        "type": "flat",
        "price": 6500000,
        "area_sqft": 1100,
        "location": {
            "type": "Point",
            "coordinates": [85.8199, 20.2961]  # lng, lat
        },
        "city": "Bhubaneswar",
        "created_at": datetime.utcnow()
    },
    {
        "title": "3BHK Patia",
        "type": "flat",
        "price": 7800000,
        "area_sqft": 1350,
        "location": {
            "type": "Point",
            "coordinates": [85.8205, 20.2972]
        },
        "city": "Bhubaneswar",
        "created_at": datetime.utcnow()
    }
]

result = db.properties.insert_many(properties)

print("Inserted property IDs:")
for _id in result.inserted_ids:
    print(str(_id))
```

```python
transactions = [
    {
        "property_type": "flat",
        "price": 6200000,
        "area_sqft": 1100,
        "location": {
            "type": "Point",
            "coordinates": [85.8201, 20.2965]
        },
        "registered_on": datetime.utcnow()
    },
    {
        "property_type": "flat",
        "price": 6000000,
        "area_sqft": 1050,
        "location": {
            "type": "Point",
            "coordinates": [85.8195, 20.2959]
        },
        "registered_on": datetime.utcnow()
    },
    {
        "property_type": "flat",
        "price": 6400000,
        "area_sqft": 1150,
        "location": {
            "type": "Point",
            "coordinates": [85.8210, 20.2970]
        },
        "registered_on": datetime.utcnow()
    }
]

db.transactions.insert_many(transactions)

print("Inserted transactions:", len(transactions))
```
